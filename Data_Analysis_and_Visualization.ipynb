```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")

# Data Set Selection
# For this example, we'll use the Titanic dataset from Kaggle
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

# Data Loading
df = pd.read_csv(url)

# Data Exploration
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())

# Display the structure of the dataset
print("\nStructure of the dataset:")
print(df.info())

# Display statistical summary of the dataset
print("\nStatistical summary of the dataset:")
print(df.describe(include='all'))

# Data Cleaning
# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Handle missing values
# Fill missing 'Age' values with the median age
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill missing 'Embarked' values with the mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop 'Cabin' column due to many missing values
df.drop(columns=['Cabin'], inplace=True)

# Check for duplicates and remove them
df.drop_duplicates(inplace=True)

# Data Visualization
# 1. Bar plot of the number of passengers embarked from each port
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Embarked', palette='Set2')
plt.title('Number of Passengers Embarked from Each Port')
plt.xlabel('Port of Embarkation')
plt.ylabel('Number of Passengers')
plt.show()
# Insight: Most passengers embarked from port 'S'.

# 2. Distribution of ages of passengers
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
# Insight: The majority of passengers were between 20 and 40 years old.

# 3. Bar plot of survival rate by sex
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Sex', y='Survived', palette='Set1')
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
# Insight: Females had a higher survival rate compared to males.

# 4. Count plot of passengers by class
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Pclass', palette='Set3')
plt.title('Number of Passengers by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.show()
# Insight: Most passengers were in the 3rd class.

# 5. Box plot of passenger age by class
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Pclass', y='Age', palette='Set2')
plt.title('Passenger Age by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
# Insight: Passengers in the 1st class were generally older compared to those in 2nd and 3rd classes.

# 6. Heatmap of the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
# Insight: There is a moderate positive correlation between 'Fare' and 'Survived', and a negative correlation between 'Pclass' and 'Survived'.

# Analysis and Insights
# The above visualizations provide several insights about the Titanic dataset:
# 1. The majority of passengers embarked from port 'S'.
# 2. The age distribution of passengers shows that most were between 20 and 40 years old.
# 3. Female passengers had a higher survival rate compared to male passengers.
# 4. Most passengers were in the 3rd class.
# 5. Passengers in the 1st class were generally older than those in other classes.
# 6. There are notable correlations between 'Fare', 'Pclass', and 'Survived'.

# The visualizations help to understand the demographic and socio-economic characteristics of the passengers and how these factors influenced their survival rates.
```
